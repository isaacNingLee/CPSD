```python
import os
import pandas as pd
import regex as re
```

```python
path = '/home/ilee0022/cl-gen/CPSD/output/20240801-130508_cp_cpsd_cifar10_anneal_mean_init_ad2/gen_samples'

dirs = os.listdir(path)

path_arr = []
counter = 0

for d in dirs:

    # if contain class_9
    if re.search('class_9', d):

        path_arr.append(d)
        counter += 1
new_df = pd.DataFrame.from_dict({'file_name': path_arr, 'label': [9]*counter, 'text':['truck']*counter, 'ucg': [0]*counter})
ori_df = pd.read_csv(path + '/metadata.csv')
        
```

```python
df = pd.concat([ori_df, new_df], ignore_index=True)
```
